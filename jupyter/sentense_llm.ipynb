### ë°ì´í„° ì „ì²˜ë¦¬ í¬í•¨ X
- ìœ íŠœë¸Œ API í˜¸ì¶œ ë°ì´í„°ì´ê¸°ì— ê²°ì¸¡ì¹˜ X
- `Comment` ì „ì²˜ë¦¬ë¥¼ ì§„í–‰í•˜ì§€ ì•ŠìŒ. (llm)

```python
import os
from openai import OpenAI
import time

# OpenAI í´ë¼ì´ì–¸íŠ¸ ì„¤ì •
client = OpenAI(
    api_key='API_KEY'
)
```

```python
# ë¶„ë¥˜ í•¨ìˆ˜
def extract_keywords(review_text, max_tokens=100):
    prompt = f"KIAíŒ¬ì„ ë¶„ë¥˜í•˜ëŠ” ì‘ì—…ì„ ì§„í–‰í•©ë‹ˆë‹¤. ë‹¤ìŒ ìœ íŠœë¸Œ ëŒ“ê¸€ í…ìŠ¤íŠ¸ë¥¼ ê¸°ë°˜ìœ¼ë¡œ í•´ë‹¹ ëŒ“ê¸€ì„ ê¸ì •, ë¶€ì •, ì¤‘ë¦½ ì¤‘ í•˜ë‚˜ë¡œ ê°ì • ë¶„ë¥˜í•´ì£¼ì„¸ìš”. ëŒ€ë‹µì€ ê¸ì •ì´ë©´ KIA, ë¶€ì •ì´ë©´ other, ì¤‘ë¦½ì´ë©´ neutralityìœ¼ë¡œ í•©ë‹ˆë‹¤.:\n\n{review_text}"
    
    try:
        # ChatCompletionì„ ì‚¬ìš©í•˜ì—¬ GPT-4 ëª¨ë¸ì— ìš”ì²­
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            max_tokens=max_tokens,
            temperature=0.5,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
        # ì‘ë‹µì—ì„œ í‚¤ì›Œë“œë¥¼ ì¶”ì¶œí•˜ì—¬ ë°˜í™˜
        keywords = response.choices[0].message.content.strip()
        return keywords
    except Exception as e:
        # ì˜ˆì™¸ ì²˜ë¦¬: ì˜¤ë¥˜ ë°œìƒ ì‹œ ë©”ì‹œì§€ ì¶œë ¥
        print(f"API í˜¸ì¶œ ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return None
```

### 01. KIA ëŒ“ê¸€ ê°ì • ë¶„ë¥˜

```python
# ì§„í–‰ ìƒí™© ì¶”ì  ë³€ìˆ˜ ì´ˆê¸°í™”
total_reviews = len(kia_data)
start_time = time.time()
```

```python
# ê° êµ¬ë‹¨ì˜ ëŒ“ê¸€ ê°ì •(ê¸ì •, ë¶€ì •, ì¤‘ë¦½) ë¶„ë¥˜
extracted_sentiment = []
for idx, row in kia_data.iterrows():
    # í‚¤ì›Œë“œ ì¶”ì¶œ
    keywords = extract_keywords(row['Comment'])
    extracted_sentiment.append(keywords)

    # ì§„í–‰ ìƒí™© ì¶œë ¥
    if (idx + 1) % 10 == 0 or (idx + 1) == total_reviews:
        elapsed_time = time.time() - start_time
        avg_time_per_review = elapsed_time / (idx + 1)
        remaining_reviews = total_reviews - (idx + 1)
        estimated_time_remaining = remaining_reviews * avg_time_per_review
        print(f"{idx + 1}/{total_reviews} ì™„ë£Œ - ì˜ˆìƒ ë‚¨ì€ ì‹œê°„: {estimated_time_remaining:.2f}ì´ˆ")
```

```python
# ê²°ê³¼ë¥¼ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
kia_data['extracted_sentiment'] = extracted_sentiment
```
