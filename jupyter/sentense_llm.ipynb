### 데이터 전처리 포함 X
- 유튜브 API 호출 데이터이기에 결측치 X
- `Comment` 전처리를 진행하지 않음. (llm)

```python
import os
from openai import OpenAI
import time

# OpenAI 클라이언트 설정
client = OpenAI(
    api_key='API_KEY'
)
```

```python
# 분류 함수
def extract_keywords(review_text, max_tokens=100):
    prompt = f"KIA팬을 분류하는 작업을 진행합니다. 다음 유튜브 댓글 텍스트를 기반으로 해당 댓글을 긍정, 부정, 중립 중 하나로 감정 분류해주세요. 대답은 긍정이면 KIA, 부정이면 other, 중립이면 neutrality으로 합니다.:\n\n{review_text}"
    
    try:
        # ChatCompletion을 사용하여 GPT-4 모델에 요청
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            max_tokens=max_tokens,
            temperature=0.5,
            top_p=1.0,
            frequency_penalty=0.0,
            presence_penalty=0.0
        )
        # 응답에서 키워드를 추출하여 반환
        keywords = response.choices[0].message.content.strip()
        return keywords
    except Exception as e:
        # 예외 처리: 오류 발생 시 메시지 출력
        print(f"API 호출 중 오류 발생: {e}")
        return None
```

### 01. KIA 댓글 감정 분류

```python
# 진행 상황 추적 변수 초기화
total_reviews = len(kia_data)
start_time = time.time()
```

```python
# 각 구단의 댓글 감정(긍정, 부정, 중립) 분류
extracted_sentiment = []
for idx, row in kia_data.iterrows():
    # 키워드 추출
    keywords = extract_keywords(row['Comment'])
    extracted_sentiment.append(keywords)

    # 진행 상황 출력
    if (idx + 1) % 10 == 0 or (idx + 1) == total_reviews:
        elapsed_time = time.time() - start_time
        avg_time_per_review = elapsed_time / (idx + 1)
        remaining_reviews = total_reviews - (idx + 1)
        estimated_time_remaining = remaining_reviews * avg_time_per_review
        print(f"{idx + 1}/{total_reviews} 완료 - 예상 남은 시간: {estimated_time_remaining:.2f}초")
```

```python
# 결과를 데이터프레임에 추가
kia_data['extracted_sentiment'] = extracted_sentiment
```

```python
# 각 구단의 댓글 감정(긍정, 부정, 중립) 분류
extracted_sentiment = []
for idx, row in samsung_data.iterrows():
    # 키워드 추출
    keywords = extract_keywords(row['Comment'])
    extracted_sentiment.append(keywords)

    # 진행 상황 출력
    if (idx + 1) % 10 == 0 or (idx + 1) == total_reviews:
        elapsed_time = time.time() - start_time
        avg_time_per_review = elapsed_time / (idx + 1)
        remaining_reviews = total_reviews - (idx + 1)
        estimated_time_remaining = remaining_reviews * avg_time_per_review
        print(f"{idx + 1}/{total_reviews} 완료 - 예상 남은 시간: {estimated_time_remaining:.2f}초")
```

```python
samsung_data['extracted_sentiment'] = extracted_sentiment
```

```python
# '긍정' 포함된 값을 'SAMSUNG'으로 변환
samsung_data['extracted_sentiment'] = samsung_data['extracted_sentiment'].apply(lambda x: 'SAMSUNG' if '긍정' in x else x)
samsung_data['extracted_sentiment'] = samsung_data['extracted_sentiment'].apply(lambda x: 'SAMSUNG' if '부정' in x else x)
samsung_data['extracted_sentiment'] = samsung_data['extracted_sentiment'].apply(lambda x: 'SAMSUNG' if '중립' in x else x)
```
