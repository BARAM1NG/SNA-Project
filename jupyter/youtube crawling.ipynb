# 기아타이거즈 크롤링

```python
# 채널 ID 찾기
search_response = youtube.search().list(
    part='snippet',
    q='kiatigerstv',  # '@' 기호 없이 채널 이름을 입력하세요
    type='channel',
    maxResults=1
).execute()

print(search_response)

# UCKp8knO8a6tSI1oaLjfd9XA
```

```python
# 특정 조건 search
'''
order의 종류
date: 최근 순으로 정렬
rating: 높은 평가 순으로 정렬
relevance: 검색 쿼리(q)에 대한 관련성을 기준으로 정렬
title: 제목에 따라 문자순으로 리소스 정렬
videoCount: 업로드한 동영상이 많은 채널 내림차순 정렬
viewCount: 리소스 조회 수가 높은 항목부터 정렬
'''
# videoID 추출
search_something = youtube.search().list(
    # 'YOUR_CHANNEL_ID'를 원하는 채널 ID
    channelId='UCKp8knO8a6tSI1oaLjfd9XA',
    # part: YouTube API가 검색 결과로 반환할 정보의 범위를 지정
    part = 'snippet',
    # order: 데이터를 어떤 기준으로 조회할 것인지를 선택
    order = 'date',
    # maxResults: 검색 결과로 반환할 항목의 최대 개수를 지정
    maxResults = 50,
    # 한국시리즈 상대방 결정 일자 ~ 한국시리즈 끝난 뒤 7일 후로 날짜 제한
    # publishedAfter: 검색할 비디오가 업로드된 시작 날짜 (ISO 8601 형식, YYYY-MM-DDThh:mm:ssZ)
    publishedAfter = '2024-10-19T00:00:00Z',
    # publishedBefore: 검색할 비디오가 업로드된 종료 날짜 (ISO 8601 형식, YYYY-MM-DDThh:mm:ssZ)
    publishedBefore = '2024-11-04T23:59:59Z',
).execute()
```

```python
search_something
```

```python
# 'youtube'는 API 객체로 설정된 것이라고 가정합니다
video_ids = [
    item['id']['videoId']
    for item in search_something['items']
    if item['id']['kind'] == 'youtube#video' and 'videoId' in item['id']
]
```

```python
# Video ID 리스트를 쉼표로 구분된 문자열로 변환, 여러 동영상을 한꺼번에 조회하기 위함
video_ids_str = ','.join(video_ids)

# API 호출을 통해 자세한 동영상 정보 가져오기
video_details_response = youtube.videos().list(
    part='snippet, statistics',
    id=video_ids_str
).execute()
```

```python
# 필요한 정보 추출
videos = [
    {
        # 제목
        'title': item['snippet']['title'],
        # videoId
        'videoId': item['id'],
        # 게시일
        'publishedAt': item['snippet']['publishedAt'],
        # 조회 수
        'viewCount': item['statistics'].get('viewCount', 'N/A'),
        # 좋아요 수
        'likeCount': item['statistics'].get('likeCount', 'N/A'),
        # 댓글 수
        'commentCount': item['statistics'].get('commentCount', 'N/A')
    }
    for item in video_details_response['items']
]

# DataFrame으로 변환
KIA_video_df = pd.DataFrame(videos)
```

```python
KIA_video_df
```

```python
# 댓글 수집 코드
# videoId: video_id에 입력

# your key
video_id = KIA_video_df['videoId']

# 코드를 저장하는 list
comments = list()
# Google API 객체 생성
api_obj = build('youtube', 'v3', developerKey=API_KEY)

# 각 video_id에 대해 댓글 수집
for video_id in video_ids:
    try:
        # API 호출 - 동영상 정보를 가져오기 위해 video title을 얻음
        video_response = api_obj.videos().list(
            part='snippet',
            id=video_id
        ).execute()
        video_title = video_response['items'][0]['snippet']['title'] if video_response['items'] else 'N/A'

        # API 호출 - 댓글 가져오기
        response = api_obj.commentThreads().list(
            part='snippet,replies',
            videoId=video_id,
            maxResults=100
        ).execute()

        # 모든 댓글을 수집하는 반복문
        while response:
            for item in response['items']:
                # 최상위 댓글 정보 추출
                comment = item['snippet']['topLevelComment']['snippet']
                comments.append([
                    comment['textDisplay'],
                    comment['authorDisplayName'],
                    comment['publishedAt'],
                    comment['likeCount'],
                    video_id,
                    video_title
                ])

                # 답글이 있는 경우 답글도 수집
                if item['snippet']['totalReplyCount'] > 0:
                    for reply_item in item.get('replies', {}).get('comments', []):
                        reply = reply_item['snippet']
                        comments.append([
                            reply['textDisplay'],
                            reply['authorDisplayName'],
                            reply['publishedAt'],
                            reply['likeCount'],
                            video_id,
                            video_title
                        ])

            # 다음 페이지가 있는 경우 계속해서 댓글을 수집
            if 'nextPageToken' in response:
                response = api_obj.commentThreads().list(
                    part='snippet,replies',
                    videoId=video_id,
                    pageToken=response['nextPageToken'],
                    maxResults=100
                ).execute()
            else:
                break
    except Exception as e:
        print(f"Error retrieving comments for video ID {video_id}: {e}")
```

```python
# DataFrame으로 변환
KIA_replies_df = pd.DataFrame(comments, columns=['Comment', 'Author', 'PublishedAt', 'LikeCount', 'VideoID', 'VideoTitle'])
```
